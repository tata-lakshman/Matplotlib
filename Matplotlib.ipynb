plt.plot(x,y,'g--o',label='first'): This line plots the data points defined by the x and y lists.

'g--o' is a format string that specifies the style of the plot:

'g' sets the line color to green.

'--' creates a dashed line.

'o' adds circular markers at each data point.

label='first' assigns a label to this line, which will be used in the legend.

plt.legend(): This line displays the legend on the plot, showing the label assigned to the line.

plt.show(): This line renders the plot and displays it.

```python
plt.plot(x,y,'g--o',label='first')
plt.legend()
plt.show()
```

```python
plt.plot(x,y,'g--o',label='first')
plt.legend(loc=2)#It is used to represent the legent in right or left(2 is left and 1 is right)
plt.show()
```

```python
plt.plot(x,y,'g--o',label='first')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.plot(x,y,'g--o',label='first')
plt.legend(loc='best')
plt.show()
```

```python
plt.plot(x,y,'r--*')
plt.grid()
plt.show()
```

"Univariate" means dealing with or analyzing one variable at a time.

In data analysis, a univariate analysis focuses on describing, summarizing, or examining the distribution of a single characteristic of your data. You've already performed univariate analysis on the "Sales" column by creating a histogram, which shows the distribution of sales values.

Common univariate visualizations include:

Histograms: Show the frequency distribution of a numerical variable.

Box plots: Display the distribution based on quartiles, showing median, spread, and potential outliers.

Density plots: Similar to histograms but use a smooth curve to show the distribution.

Bar charts: Can be used for categorical variables to show the frequency of each category.



"Bivariate" refers to the relationship or analysis of two variables.

In the context of data analysis and visualization, a bivariate analysis involves looking at how two different variables relate to each other. For example, you might want to see if there's a relationship between:

Sales and advertising cost

Temperature and ice cream sales

Years of experience and salary

Visualizations used for bivariate analysis often show how changes in one variable correspond to changes in the other. Scatter plots are a common type of bivariate visualization, where you plot data points for two variables on the same graph to see if there's a pattern or correlation.

```python
fig,axs=plt.subplots(1,3,figsize=(10,5))
#line plot
axs[0].plot(sort_age["Age"],df["Salarys"],marker='*',linestyle='--',color='green',linewidth='2')
axs[0].grid(axis='both',linestyle=':')
axs[0].set_title("Line Plot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")
#histogram
axs[1].hist(df["Salarys"],bins=10,color="pink")
axs[1].set_title("Histogram")
axs[1].set_xlabel("Salary")
axs[1].set_ylabel("Frequency")
#boxplot
axs[2].boxplot(df["Salarys"],vert=True,patch_artist=True,boxprops=dict(facecolor='red'),medianprops=dict(color='blue'),whiskerprops=dict(color='green'),capprops=dict(color='yellow'),flierprops=dict(color='pink'))
axs[2].set_title("Box Plot")
axs[2].set_xlabel("Salary")
axs[2].set_ylabel("Frequency")
plt.savefig("multipleroots.png")
plt.show()
```

```python
data={
    'year':[2020,2021,2022,2023],
    'sales':[1000,2300,1300,3000],
    'profit':[10,23,13,30],
    'expenses':[100,230,130,300]
}
df=pd.DataFrame(data)
plt.plot(df['year'],df['sales'],label='sales')
plt.plot(df['year'],df['profit'],label='profit')
plt.plot(df['year'],df['expenses'],label='expenses')
plt.xlabel('year')
plt.ylabel('amount')
plt.title('financial report')
plt.legend()
plt.show()

```

```python
#3d plot
ax=plt.axes(projection='3d')
ax.plot3D(df['year'],df['sales'],df['profit'])
plt.show()
```

```python
ax=plt.axes(projection='3d')
ax.plot3D(df['year'],df['sales'],df['profit'],color='red')
ax.set_xlabel('year')
ax.set_ylabel('sales')
ax.set_zlabel('profit')
plt.show()
```
